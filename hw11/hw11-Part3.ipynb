#### Exercise 6 -  Pie Plots

Consider the following distribution of populations of 10 countries in the European Union in 2016 (See https://en.wikipedia.org/wiki/List_of_European_Union_member_states_by_population)

```
1	 Germany	82,301,678	
2	 France	66,991,000	
3	 Italy	60,795,612	
4	 Spain	46,468,102	
5	 Poland	38,567,614	
6	 Romania	19,861,408	
7	 Netherlands	17,100,475	
8	 Belgium	11,258,434	
9	 Greece	10,812,46
10	 Czech Republic	10,538,275
```

1. Create a pie-chart that shows the distribution of populations across the 10 countries
2. Comment on your findings
3. Why could a pie-chart be problematic when trying to visualize the results

````python
### YOUR CODE GOES

import matplotlib.pyplot as plt

country = ["Germany", "France", "Italy", "Spain", "Poland", "Romania", "Netherlands", "Belgium", "Greece", "Czech Republic"]
population = [82301678, 66991000, 60795612, 46468102, 38567614, 19861408, 17100475, 11258434, 10812460, 10538275]

fig, subplots = plt.subplots(1, 2, figsize=(22, 10))
fig.suptitle("Population of European Union in 2016", color="purple", fontsize="24",fontweight="bold")
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12

subplots[0].pie(population, labels=country, shadow=True, autopct = '%1.1f%%')
subplots[0].set_title('Pie Chart', color="darkblue", fontsize="18")

subplots[1].barh(country, population, color="lightblue")
subplots[1].set_title('Horizontal Bar Chart', color="darkblue", fontsize="18")
subplots[1].set_xlabel('Population in 10s of Millions', color="blue")

plt.show()

print "Comment: The larger, more developed countries have the denser population."
print "         The problem with the pie chart is when there are more then 5 labels the chart will look crowded." 

### END CODE
````
