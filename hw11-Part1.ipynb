** Exercise 2 - Line chart **

Consider the following dataset of the weights of 7 cars and their corresponding fuel consumption

```
weight = [3170,3455,3222,3983,2441,2500,2390]
fuel_consumption (miles per gallon) = [26,30,28,23,36,33,38]
```

1. Create a line chart that plots the weight of cars against their fuel consumption.
2. Comment on your findings. How does fuel consumption change as the weight of the car increases?

Make sure to add labels and a title to your chart

````python
### YOUR CODE GOES BELOW

import matplotlib.pyplot as plt

weightFuel = []

weight = [3170,3455,3222,3983,2441,2500,2390]
fuel_consumption = [26,30,28,23,36,33,38]

# Copy weight and corresponding fuel consumption figures in a pair to a list so it could be sorted and not get mixed up.
for a in range(len(weight)):   
    weightFuel.append([weight[a], fuel_consumption[a]]) 

# Sort by weight
weightFuel.sort()

# Extract the weight list and fuel list after sorting
wgt = []
fuel = []
for a in range(len(weight)):
    wgt.append(weightFuel[a][0])
    fuel.append(weightFuel[a][1])

print "Car weight and corresponding fuel consumption sorted:"
print wgt
print fuel

plt.figure(figsize=(14, 8))
  
plt.xlabel("Weight of cars", color="blue")
plt.tick_params(axis='x', colors='red')
plt.ylabel("Fuel comsumption (Miles per gallon)", color="blue")
plt.tick_params(axis='y', colors='green')
plt.title("Weight of Cars/Fuel Consumption", color="navy", fontsize="18",fontweight="bold")
plt.plot(wgt, fuel, "bo-", color="magenta")

plt.show()

print
print "Comment: This chart tells me that most of the time, the heavier the car the less fuel it consumes. Which doesn't make sense but that's what the data says. That led me to believe that the figures must be wrong."


### END CODE
````

#### Exercise 3 - Scatter Plot

Consider the following dataset of the weights of individuals and the number of times they used the remote contorl in a period of one hour

```
weight = [121,127,128,131,130,131,133,141,161,167,169,171,173,175,177,181]
remote_use = [4,3,7,5,3,8,3,4,22,21,17,23,19,22,18,23]
```

1. Create a scatterplot with the weight on the y-axis and the remote_use variable on the x_axis.
2. What can you conclude from the weight the points are distributed on the graph? Does a particular pattern standout?

Make sure to add labels and a title to your chart


````python
### YOUR CODE GOES BELOW

import matplotlib.pyplot as plt

weight = [121,127,128,131,130,131,133,141,161,167,169,171,173,175,177,181]
remote_use = [4,3,7,5,3,8,3,4,22,21,17,23,19,22,18,23]

plt.figure(figsize=(14, 8))

plt.xlabel("Body Weight", color="green")
plt.ylabel("Remote Control Usage/Hour", color="green")
plt.title("Body Weight/Remote Control Usage per Hour", color="darkgreen", fontsize="16",fontweight="bold")
plt.scatter(remote_use, weight, edgecolors="red", color="orange")

plt.show()

print "Comment: This shows that the heavier the body weight the less the person is willing to move."

### END CODE
````
